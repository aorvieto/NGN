```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('results/res18_cifar10_ngnm_s0_lr3.0_decayFalse_uid18e8a3174a.csv')

# Extract the 'lr' column
lr_values = df['lr']

# Plot the 'lr' values
plt.plot(range(1,len(lr_values)+1),lr_values)

# Add labels and title
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate over Epochs')
plt.xscale('log')
# Show the plot
plt.show()
```
